The **Titanic dataset** in Seaborn (`sns.load_dataset("titanic")`) contains data on Titanic passengers, including details about their survival, demographics, and ticket information.

### **Titanic Dataset Columns**

| **Column** | **Description** |
|------------|------------------|
| `survived` | Survival status (0 = No, 1 = Yes) |
| `pclass`   | Passenger class (1 = First, 2 = Second, 3 = Third) |
| `sex`      | Gender of the passenger (`male` or `female`) |
| `age`      | Age of the passenger in years (some values are missing) |
| `sibsp`    | Number of siblings or spouses aboard the Titanic |
| `parch`    | Number of parents or children aboard the Titanic |
| `fare`     | Passenger fare (in British pounds) |
| `embarked` | Port of embarkation (`C` = Cherbourg, `Q` = Queenstown, `S` = Southampton) |
| `class`    | Passenger class (categorical version of `pclass`) |
| `who`      | Type of passenger (`man`, `woman`, or `child`) |
| `deck`     | Deck level (`A` to `G`, `NaN` if unknown) |
| `embark_town` | Port of embarkation (`Cherbourg`, `Queenstown`, `Southampton`) |
| `alive`    | Survival status (`yes` or `no`) |
| `alone`    | Whether the passenger was alone on the ship (`True` or `False`) |

### **Additional Notes:**
- **`sibsp` and `parch`** help determine family size onboard.
- **`class` and `pclass`** represent the same information but in different formats (categorical vs numerical).
- Missing values (`NaN`) in columns like `age` and `deck` represent unknown information.

```python
Unique = df.groupby(['embark_town','survived'])['age'].unique()
print(Unique)
```

```python
mean_age_df = df.groupby(['embark_town', 'survived'])['age'].mean().reset_index()
sns.barplot(data=mean_age_df, x='embark_town', y='age', hue='survived')
```

```python
count_df = df.groupby(['embark_town', 'survived'])['age'].count().reset_index()
sns.barplot(data=count_df, x='embark_town', y='age', hue='survived')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the countplot with custom colors
sns.countplot(x='survived', hue='sex', data=df, palette='coolwarm')
plt.title('Count of People by Survival Status')
plt.xlabel('Survival Status (0 = Did not survive, 1 = Survived)')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping the data and counting the occurrences
grouped_data = df.groupby(['survived', 'sex', 'who'])['who'].count().reset_index(name='count')

# Plotting the countplot
sns.countplot(x='survived', hue='who', data=grouped_data, palette='Set2')
plt.title('Count of People by Survival Status, Who, and Sex')
plt.xlabel('Survival Status (0 = Did not survive, 1 = Survived)')
plt.ylabel('Count')
plt.show()
```
