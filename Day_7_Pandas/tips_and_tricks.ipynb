# 19-Reshaping Mutli_Index Series

```python
df.groupby('sex').survived.mean()
```

```python
df.survived.mean()
```

```python
df.groupby(['sex', 'class'], observed=True).survived.mean().unstack()
```

```python
df.groupby(['sex', 'class'], observed=True).survived.mean()
```

```python
df.groupby(['who','class'], observed=False).survived.mean()
```

```python
import pandas as pd

# Data with missing combination (no 'male' in 'Third')
data = {
    'sex': pd.Categorical(['female', 'female', 'male']),
    'class': pd.Categorical(['First', 'Second', 'First']),
    'survived': [1, 0, 1]
}
df_1 = pd.DataFrame(data)

# With observed=True (missing combinations excluded)
print(df_1.groupby(['sex', 'class'], observed=True)['survived'].mean())

# With observed=False (missing combinations included)
print(df_1.groupby(['sex', 'class'], observed=False)['survived'].mean())
```

# 23-Reshaping a DataFrame

```python
fasla = pd.DataFrame([['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]],columns=['zip','factory','warehouse','retaile'])
fasla.head()
```

```python
fasla_long = fasla.melt(id_vars='zip', var_name='location',value_name='distance')
fasla_long
```

```python
sns.barplot(x='zip', y='distance', hue='location', data=fasla_long)
```

```python
import pandas as pd

df = pd.DataFrame({
    'student': ['Ali', 'Sara'],
    'math': [90, 80],
    'english': [85, 95]
})
```

```python
df_melted = pd.melt(df, id_vars=['student'], var_name='subject', value_name='score')
```

```python
sns.barplot(data=df_melted, x='subject', y='score', hue='student')
```
