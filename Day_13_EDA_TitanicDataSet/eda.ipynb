**Note**
- Do Numeric Value ma Sy Ak Ka Group Bna Lo Using `Cut()`

```python
# Boxplot: Fare distribution across different age groups
# Load Titanic dataset if not already loaded
df_1 = sns.load_dataset('titanic')

# Remove nulls for cleaner plot
df_1 = df_1[['age', 'fare']].dropna()

# Create age bins (you can change number of bins as needed)
df_1['age_group'] = pd.cut(df_1['age'], bins=5)

# Boxplot: Fare distribution across different age groups
plt.figure(figsize=(10,6))
sns.boxplot(x='age_group', y='fare', data=df_1, palette='Set2')
plt.title('Fare Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Fare')
plt.xticks(rotation=45)
plt.show()
```
